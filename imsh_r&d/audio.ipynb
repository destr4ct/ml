```python
import librosa
import numpy as np

#   
audio_path = '/Users/vlados/Downloads/dog-bark1.wav'
y, sr = librosa.load(audio_path)

#   
f0, voiced_flag, voiced_probs = librosa.pyin(y, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7'))

#   
import matplotlib.pyplot as plt
times = librosa.times_like(f0)

plt.figure()
plt.plot(times, f0, label='f0', color='blue', linewidth=1.5)
plt.legend(loc='upper right')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Fundamental frequency (F0) estimation')
plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

#   
audio_path = '/Users/vlados/Downloads/cat-kitten2.wav'
y, sr = librosa.load(audio_path)

#   
tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)

#  
beat_times = librosa.frames_to_time(beat_frames, sr=sr)
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, alpha=0.6, color="black")
plt.vlines(beat_times, -1, 1, color='r', alpha=0.75, label='Beats')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
