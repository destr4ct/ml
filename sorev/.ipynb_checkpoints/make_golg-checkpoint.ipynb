```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('./data/make_gold_features.csv')
df.head(3)
```

```python
g = sns.relplot(
    data=df,
    x='ride_date',
    y='deviation_normal',
    style='car_id',
    aspect=4,
    hue='target_class',
    kind='line'
)
g.set_xticklabels(rotation=5, horizontalalignment='right', step=5)
```

```python
df_gold = df.groupby(['ride_date', 'car_id']).deviation_normal.mean().reset_index()
df_gold.head(10)
```

```python
df_gold['next_dat'] = df_gold.groupby('car_id').deviation_normal.shift(-1)
```

```python
df_gold['dif'] = np.abs(df_gold.next_dat - df_gold.deviation_normal)
df_gold
```

```python
result = df_gold.groupby('car_id').agg(
    gold_feature_1=('dif', 'max'),
    gold_feature_2=('dif', lambda x: x.argmax())
)
result.head()
```
