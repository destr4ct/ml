```python
import pandas as pd
```

```python
df_train = pd.read_csv('../data/car_train.csv')
df_rides = pd.read_csv('../data/rides_info.csv')

df = df_rides.merge(df_train, on='car_id')
df.fillna(0, inplace=True)

df.head(3)
```

```python
t1 = df.pivot_table(
    index='user_id',
    values=['ride_id'],
    columns=['car_type'],
    aggfunc=['count']
)

t1.columns = [f'{c[2]}_{c[0]}' for c in t1.columns]
t1.reset_index(inplace=True)
t1.fillna(0, inplace=True)
t1.head()
```

```python
t2 = df.pivot_table(
    index='user_id',
    values=['distance', 'ride_cost'],

    aggfunc=['min', 'mean', 'max']
)
t2.columns = [f'{c[1]}_{c[0]}' for c in t2.columns]
t2.reset_index(inplace=True)
t2.fillna(0, inplace=True)

t2.head()
```

```python
t1 = df.pivot_table(
    index='user_id',
    values=['ride_id'],
    columns=['car_type'],
    aggfunc=['count']
)

t1.columns = [f'{c[2]}_{c[0]}' for c in t1.columns]
t1.reset_index(inplace=True)
t1.fillna(0, inplace=True)

t2 = df.pivot_table(
    index='user_id',
    values=['distance', 'ride_cost'],

    aggfunc=['min', 'mean', 'max']
)
t2.columns = [f'{c[1]}_{c[0]}' for c in t2.columns]
t2.reset_index(inplace=True)
t2.fillna(0, inplace=True)

t3 = df.pivot_table(
    index='user_id',
    values=['car_id'],
    aggfunc=['nunique']
)
t3.columns = [f'{c[1]}_{c[0]}' for c in t3.columns]
t3.reset_index(inplace=True)
t3.fillna(0, inplace=True)


t_all = t1.merge(t2, on='user_id').merge(t3, on='user_id')
t_all['crazy_driver'] = df.speed_max.map(lambda x: int(x > 100))
t_all.head()
```
