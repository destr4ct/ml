```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://stepik.org/media/attachments/lesson/779914/make_it_easy.csv')
df.head(3)
```

```python
dft = df.copy(True)

enc = LabelEncoder()
dft['label'] = enc.fit_transform(df.label)
```

```python
sns.relplot(
    data=dft,
    x='feature_1',
    y='feature_2',
    hue='label'
)



# x**2 + y**2 < 2
```

```python
dft['feature_4'] = ((dft.feature_1**2 + dft.feature_2**2) < 2.2**2).astype(int)
sns.relplot(
    data=dft,
    x='feature_1',
    y='feature_4',
    hue='label'
)
```
